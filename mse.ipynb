```python
import numpy as np
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score
```

```python
y_actual =  np.array([10,15,2,21,9])
y_predict = np.array([12,14,10,24,12])
```

```python
mean_manual = np.mean((y_actual - y_predict) ** 2)
sk_mean = mean_squared_error(y_actual, y_predict)
```

```python
print(mean_manual)
print(sk_mean)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.scatter(range(len(y_actual)) , y_actual , label="Actual Values" , color="blue")
plt.scatter(range(len(y_predict)) , y_predict , label="Predicted Values" , color="Red")
plt.plot(range(len(y_actual)) , y_actual , linestyle="dashed" , color="blue" , alpha=0.5)
plt.plot(range(len(y_predict)) , y_predict , linestyle="dashed" , color="red" , alpha=0.5)
plt.legend()
plt.xlabel("Task Index")
plt.ylabel("Time Slot")
plt.title("Actual vs Predict")
plt.show()
```

```python
# Mean Absolute Error

## Test
```

```python
meanAbsoluteError = mean_absolute_error(y_actual , y_predict)
print("Mean Absolute Error: " , meanAbsoluteError)
```

```python
plt.figure(figsize=(8,5))
plt.plot(sk_mean  , label="MSE" , marker="o")
plt.plot(meanAbsoluteError  , label="Mean Absalute Error" , marker="s" , linestyle="dashed")
plt.axhline(0,color="black" , linewidth=0.5)
plt.legend()
plt.title("MSE vs MAE")
plt.ylabel("Data point Index")
plt.xlabel("Error")
plt.show()

```

```python
r2 = r2_score(y_actual,y_predict)

print("R2 score is: " , r2 )
```

```python
plt.plot(y_actual, label="Actual" , marker="o")
plt.plot(y_predict, label="Predictet" , marker="x")
plt.axhline(np.mean(y_actual), color="gray", linestyle="--" , label="Avarage Guess")
plt.legend()
plt.title("R² Score Example")
plt.xlabel("Task Index")
plt.ylabel("Hour")
plt.show()
```

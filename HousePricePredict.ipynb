```python
import numpy as np
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error , r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# x , y = fetch_california_housing()

data = fetch_california_housing()
x = data.data
y = data.target

# print(x)
print(y)
```

```python
features_names = data.feature_names

x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2 , random_state=42)

model = LinearRegression()
model.fit(x_train , y_train)
y_predict = model.predict(x_test)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

mse = mean_squared_error(y_test , y_predict)
r2 = r2_score(y_test , y_predict)

print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)

```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_predict, color='blue', alpha=0.6, label='Predicted vs Actual')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Ideal Prediction Line')
plt.xlabel('Actual House Value ($100,000s)')
plt.ylabel('Predicted House Value ($100,000s)')
plt.title('California Housing: Actual vs Predicted')
plt.legend()
plt.grid(True)
plt.show()
```
